### Climate Opinion Data

To start cleaning, we load the yale climate data into a pandas DataFrame:

```python
yale_climate = pd.read_csv('data/yale_climate.csv')
yale_climate.head()
```

Since we are only looking at state data, we filter the DataFrame so that only the state rows are included.

```python
yale_climate = yale_climate[yale_climate['geotype'] == 'state']
yale_climate = yale_climate[['geoname','varname', 'x2018', 'x2019', 'x2020', 'x2021', 'x2022', 'x2023']]

#consistent state names across both datasets
yale_climate['geoname'] = yale_climate['geoname'].str.upper() 
yale_climate.head()
```

The year data is originally stored as columns, but for the purpose of our study, it would make more sense to expand the data vertically and shrink it horizontally. This is done by using pandas "melt" function:

```python
melted = pd.melt(yale_climate, id_vars=['geoname', 'varname'], var_name='year', value_name='percentage')
melted['year'] = melted['year'].str.replace('x','').astype(int)
yale_climate = melted
yale_climate.head()
```

The above output shows the final version of the yale climate dataset, formatted so that it can be easily merged with the storm events data.

### Storm Data

The storm data has to be downloaded by year, so there is a csv file corresponding to each year we want to look at. We chose the range of 2018-2023, as the yale climate data has many missing values before 2018. The code below loads each csv file in and creates two DataFrames: one for event count by state, and one maintaining all of the event data.

```python
event_dfs = []
by_state_dfs = []
def load_storm(year):
    df = pd.read_csv(f'data/storm_dataset{year}.csv')
     #filters out non states
    df = df[df['STATE'].isin(state_names)] 
    #there are multiple events for the same episode, this removes them
    unique = df.drop_duplicates(subset=['STATE_FIPS', 'EPISODE_ID']) 
    agg_funcs = {
        'EVENT_ID': 'count',
        'YEAR': 'first',
    }
    by_state = unique.groupby('STATE').agg(agg_funcs).reset_index()
    by_state = by_state.rename({'EVENT_ID': 'COUNT'}, axis=1)
     #only columns we might use
    events = unique[['STATE', 'EVENT_TYPE', 'YEAR', 'EPISODE_ID', 'EPISODE_NARRATIVE',  'INJURIES_DIRECT', 'INJURIES_INDIRECT', 'DEATHS_DIRECT', 'DEATHS_INDIRECT']]
    event_dfs.append(events)
    by_state_dfs.append(by_state)
```

```python
years = [2018 + i for i in range(6)]
for year in years:
    load_storm(year)
```

```python
#note: some 'states' on here are not actual states (like Lake Superior) - we will ignore these for now
events_df = pd.concat(event_dfs, axis=0)
event_count_df = pd.concat(by_state_dfs, axis=0)
event_count_df.head()
```

The output above shows what the event count DataFrame looks like after merging all of the storm datasets into one.

```python
events_df_merged = events_df.merge(yale_climate, how='inner', left_on=['STATE', 'YEAR'], right_on=['geoname','year'])

#data for each event type along with yale climate data
events_df_merged = events_df_merged.drop(columns=['geoname', 'year']) 
events_df_merged.head()
```

```python
event_count_merged = event_count_df.merge(yale_climate, how='inner', left_on=['STATE', 'YEAR'], right_on=['geoname', 'year'])

#data with number of events for each year along with the yale climate data
event_count_merged = event_count_merged.drop(columns=['geoname', 'year']) 
```

### Change in events:
Let's also look at how the number of weather events has changed over time in each state. We'll look at the averages for the first 2 years (2018/2019) vs last 2 years (2022/2023) of our data. While the scope may be small, it gives us some sort of idea of how weather events have been changing by state the past few years.

```python
#Averages of first 2 years
event_years_first = event_count_df[event_count_df['YEAR'].isin([2018, 2019])]
event_years_first = event_years_first.groupby('STATE')['COUNT'].mean().reset_index()
```

```python
#Averages of last 2 years
event_years_second = event_count_df[event_count_df['YEAR'].isin([2022, 2023])]
event_years_second = event_years_second.groupby('STATE')['COUNT'].mean().reset_index()
```

```python
#Differences by state
merged_states = pd.merge(event_years_first, event_years_second, on='STATE', suffixes=('_df1', '_df2'))
merged_states['count_difference'] = merged_states['COUNT_df2'] - merged_states['COUNT_df1']
merged_states = merged_states.sort_values('count_difference', ascending=False)
```

We saw that the states that saw a biggest increase in events from first 2 to last 2 years were Texas, Alaska, Oklahoma, Nevada, and Florida.
According to PolicyGenius (https://www.policygenius.com/homeowners-insurance/best-and-worst-states-for-climate-change/) the states that will fare worst in the face of climate change are Florida, Mississippi, Louisiana, Texas, and Arkansas. 4 of these states are in the top 10 of increasing weather events.\
One thing to note is states like Texas that are larger experience more events due to their size, so it's also important to factor in percent change. However, it's notable that California isn't in the top 3 despite its large land area while Texas is.

```python
#percent_change
merged_states['percent_change'] = (merged_states['COUNT_df2'] - merged_states['COUNT_df1'])/merged_states['COUNT_df1']
merged_states = merged_states.sort_values('percent_change', ascending=False)
```

```python
merged_states = merged_states.sort_values('count_difference', ascending=False)
combined_df_per = pd.concat([merged_states.head(3), merged_states.tail(3)])
combined_df_per['STATE'] = combined_df_per['STATE'].apply(lambda x: x[:7])

plt.rcParams["figure.figsize"] = (10,4)
combined_df_per.reset_index(drop=True, inplace=True)
combined_df_per.plot(kind='bar', x='STATE', y='count_difference', legend=False)
plt.xlabel('State')
plt.ylabel('Count Difference')
plt.title('Bar Chart of Top 3 and Bottom 3 States')
plt.xticks(rotation=0)
plt.show()
```

Despite Texas' large area, we can see that its weather events have increased disproportionately to its size. While California is close to Texas in size, it's nowhere near it on the sorted dataframe for difference in weather events over 5 years. So, it can be inferred that Texas is being disproportionately affected by climate change and is seeing a significant increase in extreme weather events.

## Linear Regression Model for Predicting Climate Opinion

One way of understanding the relationship between extreme weather events and climate opinion is building linear regression models. These can show how well a selection of variables can predict a response variable (in this case, climate opinion).

Lets start by looking at a simple model predicting climate opinion from state event counts over the past six years:

```python
research_questions = ['affectweather', 'worried', 'regulate', 'governor', 'discuss'] 
count_models = []
for question in research_questions:
    filtered = event_count_merged[event_count_merged['varname'] == question].groupby('STATE').agg({'COUNT': 'sum', 'percentage': 'mean'})
    model = smf.ols('percentage ~ 1 + COUNT', filtered).fit()
    count_models.append(model)
    # print(question + ':')
    # print(model.summary().tables[0])
    # print('\n')
```

The results from these models show that in general, there is no significant correlation between number of events in a state and climate opinion. The lowest p-value was for 'regulate' with 0.172, which is not statistically significant. Also, all models had a slight negative coefficient, suggesting that if anything there might be a small negative correlation between number of extreme weather events and people believing in climate change/wanting the government to do more for it.

Another model we can look at is one that uses the event_df_merged, which will incorporate more variables such as event type and the number of deaths/injuries

```python
categorized = events_df_merged.copy()
categorized['EVENT_TYPE'] = pd.Categorical(categorized['EVENT_TYPE'])
```

```python
event_models = []
for question in research_questions:
    filtered = categorized[categorized['varname'] == question]
    model = smf.ols('percentage ~ EVENT_TYPE + INJURIES_DIRECT + INJURIES_INDIRECT + DEATHS_DIRECT + DEATHS_INDIRECT', filtered).fit()
    event_models.append(model)
    # print(question + ':')
    # print(model.summary().tables[0])
    # print('\n')
```

Looking through the tables from the regression model output, we can indentify some patterns. For one, the `DEATHS_DIRECT` seems to have a statistically signficant positive correlation with percentage across all questions. `INJURIES_DIRECT`, `INJURIES_INDIRECT`, and `DEATHS_INDIRECT` do not seem to have any statistically significant correlation with percentage. For the different event types, it seems that there is a lot of variance across the different types and the different research questions. Some of the coefficients are negative, although most are positive. 

Many event types have stastically significant results for one question but not all. Some types, such as Lake-Effect Snow, Coastal Flood, and Rip Current have stastically significant positive correlations across all questions. This suggests that areas that have these events more frequently might be more likely to be more conscientious about climate change.

Another pattern in these results is that some research questions have much more statistically significant results than others. For example, 'discuss' has p-values of under 0.05 for almost all variables. On the other hand, 'worried' has few p-values less than 0.05.

## Assumptions for Linear Regression

To make sure that Linear Regression is a valid method of analysis for these models, we must run some diagnostics to make sure the assumptions are met

### Normality

```python
#used to standardize residuals
def standardize(X):
    return (X - X.mean()) / X.std()
```

## Analysis of Variance (ANOVA) for Climate Opinion Model

### State & Year Covariates
Inline with previous analysis, we construct a simple linear regression model which predicts the percentage of support of climate opinon questions given the count of weather effects and the category of question. We want to test whether the inclusion of varaibles like the `STATE` and `YEAR` significantly improves the model fit. To do this, we will carry out an analysis of variance. We define the larger and smaller models to be as follows:

$$
\begin{aligned}\\
\text{Small model:} \quad \text{percent support}_i &= \beta_0 + \beta_1(\text{count of weather events}) + \beta_2(\text{question}) + \epsilon_i \\ \\
\text{Large model:} \quad \text{percent support}_i &= \beta_0 + \beta_1(\text{count of weather events}) + \beta_2(\text{question}) + \beta_3(\text{year}) + \beta_4(\text{state}) + \epsilon_i \\ \\
\end{aligned}
$$

and our test statistic, $\hat F$, to be:

$$
\hat F = {\frac{\text{SS}_{Reg}(\text{Small Model}) - \text{SS}_{Reg}(\text{Big Model})}{\Delta \text{df}} \Bigg / {\frac{\text{SS}_{Res}(\text{Big Model})}{\text{df}_{Res}(\text{Big Model})}}} \\
$$

We carry out this test at a significance level of, $\alpha = 0.01$.

```python
small_formula = f'percentage ~ COUNT + varname'
big_formula = f'percentage ~ '+ '+'.join(event_count_merged.columns.difference(['percentage']))
```

```python
small_model = smf.ols(small_formula, data=event_count_merged).fit()
big_model = smf.ols(big_formula, data=event_count_merged).fit()

anova_table = sm.stats.anova_lm(small_model, big_model)
anova_table
```

Looking at the ANOVA table, the pvalue of our test is calculated to be:

```python
anova_table['Pr(>F)'].values[1]
```

At a significance level of $\alpha = 0.01$ we come to conclusion that we reject the null hypothesis that the inclusion of the defined additional covariates does not improve model fit.**Hence, there is sufficient evidence to conclude that the addition of the covariates `YEAR` and `STATE` improves the model fit for predicting the response variable of `percentage`.**

## Hypotheis Testing

Assumption: Since our data is slightly but not incredibly skewed and we are taking a specific subset of our total data (which is a survey not a population), we can perform a t test to see if the states we deemed most impacted by climate change feel more strongly on the topic in the direction of more climate action.

$H_0$: $\hat \mu = \mu_0$ (i.e. top 5 sample mean is equal to the total survey mean for given question)\
$H_1$: $\hat \mu \gt \mu_0$ (i.e. top 5 sample mean is greater than total survery mean for given question)

```python
def t_test(question):
    yale_climate_23_gov = yale_climate[(yale_climate['year']==2023) & (yale_climate['varname']==question)]
    nat_avg = yale_climate_23_gov['percentage'].sum() / yale_climate_23_gov.shape[0]
    top5_df = yale_climate_23_gov[yale_climate_23_gov['geoname'].isin(['TEXAS', 'ALASKA', 'OKLAHOMA', 'NEVADA', 'FLORIDA'])]
    merged_states2 = merged_states.sort_values('count_difference', ascending=False)
    states = merged_states2['STATE'][0:5]
    mean_gov_states = 0.0
    for i in states:
        mean_gov_states += yale_climate_23_gov[yale_climate_23_gov['geoname']==i]['percentage'].iloc[0]
    mean_gov_states = mean_gov_states/5
    Xbar = mean_gov_states
    mu = nat_avg
    stdev = np.std(yale_climate_23_gov['percentage'])
    stdev_hat = top5_df['percentage'].std()*np.sqrt(5)/np.sqrt(4) #std uses denominator of sqrt(n), we need sqrt(n-1)
    n = 5
    t_n = (Xbar-mu)/(stdev_hat/np.sqrt(n))
    pval = 1 - t.cdf(t_n, n-1)
    print('Question topic: ' + question) 
    print('\nXbar (top 5 mean): ' + str(Xbar))
    print('Mu (National Avg): '+ str(nat_avg))
    print('T stat: '+ str(t_n))
    print('p value: ' + str(pval))
    if(pval>0.05):
        print('\np-value high, fail to reject null hypothesis: no significant relation between weather events and this question')
    else:
        print('\np-value low, reject null hypothesis')
    return [t_n, stdev_hat, mu, stdev]
```

```python
def t_hist(question):
    fig, axs = plt.subplots(1,1, figsize=(6,2))
    datas = t_test(question)
    plt.hist(yale_climate[(yale_climate['year']==2023) & (yale_climate['varname']==question)]['percentage'], density=True)
    x = np.linspace(20, 90, 1000)
    scale = datas[3]
    y = t.pdf((x - datas[2]) / scale, 50) * (1 / scale)
    plt.plot(x, y, 'r-', lw=2, label=f't-distribution centered at {datas[2]} (df=50)')
    plt.xlabel("percentage answered yes to question")
    plt.ylabel("propoortion")
    plt.vlines(datas[0] * datas[1]/np.sqrt(5) + datas[2], 0, 0.15, color='red', linestyles='--', lw=2)
```

# Discussion and Conclusion

The linear regression analysis generally yielded statistically insignificant results. The models using event count by state as the predictor all resulted in models with p-values greater than 0.17, suggesting that there is likely no relationship between number of extreme weather events and climate opinion. Additionally, the coefficients for these regression models were almost zero, furthering the idea that the existence of a relationship between these two variables lacks evidence. The second set of linear regression models were more complex, as they used event type, direct/indirect deaths, and direct/indirect injuries. However, the assumptions for linear regression were not met for the more complex set of models, meaning no inferences can be made from the results of these models.

From conducting further tests with ANOVA the results show that for a model that uses frequency of weather events to predict percentage support of responses to climate opinion questions, the state and recorded year improve the fit of the model. This might signify that geographical location has impacts on the outcome of surveyed climate opinions yet further study would be needed to explore this more concretely. Evidence of improving model fit is also true for casualty covariates in a model with percent support once again as the response variable. This perhaps entails that the severity of an extreme weather event may factor to opinion on climate change. However, to truly understands if there is evidence of this in underlying data, further testing and seperate studies must be conducted. 

When conducting our t-tests on the 5 states with the biggest increases in extreme weather and comparing their answers to the full country, we found that of the 5 questions we examined, our subset was not significantly shifted towards the pro climate action side on any. In summary, there is no significant evidence that people who are experiencing worsening weather (mostly due to climate change) are more concerned or more in favor of action being taken than the country as a whole. In fact, in many cases our subset average proportion was less than the national average. There are several factors that may account for this. One is that most of our selected states lean Republican, and typically red states are less concerned about the issue of climate change and environmental regulation. However, it is still very interesting that the states that seem to be faring worst in the face of a climate crisis are actually less concerned about the issue on average than other states. The two questions where the subset was close to or to the right of the mean were "are you worried about climate change?" and "do you think the governor should be doing more about climate change?". There seems to be some sort of concern about the issue and hope that the government does something to address it, but it's not significantly more than states that aren't facing the same increase in extreme weather events.

In conclusion, when taking into account all conducted analysis overall there is not sufficient statistical evidence to conclude that there exists a relationship between climate change opinons and extreme weather events. Hence, our hypothesis is not proven to be true as we cannot infer if the frequency of extreme weather events in a location has an effect on climate opinion in that area. Certain limitations exist in this study due to the datasets we used and the methods of analysis we employed. For one, looking on a state level loses variability in weather that would be seen if looking on a county or zip code level. Many states have good variety in weather depending on location, but this is lost when looking at a state level. Another limitation is using event frequency as a predictor. This is due to it not accounting for the severity of different events, as some could be much more impactful than others. Additionally, there are potential confounding factors that we did not account for in our analysis. This could include variables such as political leaning or age. Confirmation bias may also be attributible to why we found little to no connection between response and extreme weather experience. In an age where you can get your information from cherry-picked sources, we see a lot more of this phenomenon among populations. Even if evidence suggests that climate change is real and an imminent threat, if you're consuming media that frames the evidence in a way that simply reinforces your pre-existing beliefs, your real-world experience may play less of a role in forming your opinion. Since we've found extreme weather events to be a non-factor in climate opinion, it would be interesting to see how these other variables play a role in what people believe about the climate crisis.
